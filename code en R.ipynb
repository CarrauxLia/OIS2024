```r
# Installation et chargement des packages nécessaires
install.packages("ggplot2")
library(ggplot2)

# Création d'un jeu de données fictif
set.seed(123)
house_data <- data.frame(
  surface = rnorm(100, mean = 2000, sd = 500),
  price = rnorm(100, mean = 300, sd = 100),
  rooms = sample(2:5, 100, replace = TRUE)
)

head(house_data)

# Création du graphique
p <- ggplot(house_data, aes(x = surface, y = price)) +
  geom_point(aes(color = as.factor(rooms)), size = 3) + 
  geom_smooth(method = "lm", se = FALSE, color = "black") + 
  facet_wrap(~ rooms) + 
  labs(title = "Relation entre la surface habitable et le prix des maisons",
       x = "Surface habitable (pieds carrés)",
       y = "Prix (en milliers de dollars)",
       color = "Nombre de chambres") +
  theme_minimal() 

# Affichage du graphique
print(p)
```
